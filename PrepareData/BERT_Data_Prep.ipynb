## Dialogue Acts - Sentence by Sentence

```python
def condenseDA(da = ''):
    # Condense the Dialogue Acts Classes
    
    #Segmentation = Fragment + Backchannel + Stall
    if da == '1' or da == '2' or da == '3': 
        da = '0'
    
    #Information = Inform + Elicit Inform
    elif da == '4' or da == '5': 
        da = '1'
    
    #Actions = Suggest + Offer + Elicit Offer or Suggestion
    elif da == '6' or da == '7' or da == '8': 
        da = '1'
    
    #Discussion = Assessment + Comment about Understanding + Elicit Assessment + Elicit Comment about Understanding
    elif da == '9' or da == '11' or da == '12' or da == '13': 
        da = '1'
    
    #Social = Be Positive + Be Negative
    elif da == '14' or da == '15': 
        da = '0'
    
    #Other
    elif da == '16': 
        da = '0'
    
    return da
```

```python
def prepareBertFile(bertfile, path, rem1word = False):
    lineid = 1
    with open(bertfile, mode='w', newline='') as tsvFile:
        writer = csv.writer(tsvFile, delimiter='\t')
        
        if bertfile == 'test.tsv':
            flag = True
            writer.writerow(['id', 'sentence'])
        else:
            flag = False
            
        for filename in os.listdir(path):
            if flag:
                writer.writerow([lineid, 'START OF NEW MEETING'])
            else:
                writer.writerow([lineid, '0', 'a', 'START OF NEW MEETING'])
            lineid += 1
            file = open(path + filename, "r")
            tempLine = []
            da = ''
            for line in file:
                tempLine = line.split()[2:-1]
                
                if rem1word and len(tempLine) <= 3:
                    continue
                
                sent = ' '.join(tempLine)
                da = condenseDA(line.split()[-1].split('_')[2])
#                 sents = wrap(tempLine, 125)
#                 for sent in sents:
                try:
                    if flag:
#                         print([lineid, sent])
                        writer.writerow([lineid, sent])
                    else:
#                         print([lineid, da, 'a', sent])
                        writer.writerow([lineid, da, 'a', sent])
                except:
                    pass
                lineid += 1
#             if flag:
#                 writer.writerow([lineid, ' '])
#             else:
#                 writer.writerow([lineid, '0', 'a', ' '])
            if flag:
                writer.writerow([lineid, 'END OF MEETING'])
            else:
                writer.writerow([lineid, '0', 'a', 'END OF MEETING'])
            lineid += 1
    print('{} file saved successfully'.format(bertfile))
```
